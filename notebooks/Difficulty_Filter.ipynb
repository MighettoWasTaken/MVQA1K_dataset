```python
from smart_open import open as smart_open
from PIL import Image
import io
import pandas as pd
import gcsfs


train_path = 'gs://cse242_vqa/PMC-VQA/train.csv'
test_path = 'gs://cse242_vqa/PMC-VQA/test.csv'
def load_image(path):
    local_dir = '/content/images/images/'
    with open(local_dir + path, 'rb') as f:
        return Image.open(io.BytesIO(f.read())).convert("RGB")


def load_csv(csv_path):
  with smart_open(csv_path, "r") as f:
    return pd.read_csv(f)

train_file = load_csv(train_path)
test_file = load_csv(test_path)
```

```python
from tqdm import tqdm
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
import torchvision.transforms as T

import torch
torch.cuda.empty_cache()
# Detect available device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"device: {device}")


# Move the model to the detected device
model.to(device)

preprocess = T.Compose([
    T.Resize((448, 448)),
    T.ToTensor(),
    T.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
])

autosave_path = 'partial_results.csv'
save_interval = 500
prev_end = 59000
embedder = SentenceTransformer('all-MiniLM-L6-v2')
batch_size = 32 # Tune this based on GPU memory
results = []

batch = []
write_header = not os.path.exists(autosave_path)
def process_batch(batch):
    batch_res = []
    images = [load_image(row['Figure_path']) for row in batch]
    images = torch.stack([preprocess(img) for img in images])
    questions = [row['Question'] for row in batch]
    all_choices = [[row['Choice A'], row['Choice B'], row['Choice C'], row['Choice D']] for row in batch]
    correct_answers = [row['Answer'] for row in batch]

    # Format prompt text for each question
    texts = [
        f"[INST]<ImageHere>\nQuestion: {q}\nChoices: \nA. {c[0]} \nB. {c[1]} \nC. {c[2]} \nD. {c[3]}\n Repeat full answer[/INST]"
        for q, c in zip(questions, all_choices)
    ]

    # Batched model inference
    model_outputs = model.generate(images=images, texts=texts)

    # Embed everything at once
    flat_texts = []
    index_map = []  # (start_idx in flat list, length)

    for out, chs, ans in zip(model_outputs, all_choices, correct_answers):
        group = [out] + chs + [ans]
        index_map.append((len(flat_texts), len(group)))
        flat_texts.extend(group)

    embeddings = embedder.encode(flat_texts, convert_to_tensor=True)

    # Organize results
    for row, (start, count) in zip(batch, index_map):
        group_embed = embeddings[start:start + count].to("cpu")
        model_out_embed = group_embed[0:1]
        choices_embed = group_embed[1:5]
        answer_embed = group_embed[5:6]

        sim_to_choices = cosine_similarity(model_out_embed, choices_embed)[0]
        sim_to_answer = cosine_similarity(model_out_embed, answer_embed)[0][0]


        batch_res.append({
            'figure_path': row['Figure_path'],
            'question': row['Question'],
            'model_output': flat_texts[start],
            'true_answer': row['Answer'],
            'similarity_to_answer': sim_to_answer,
            'max_similarity_to_choices': sim_to_choices.max().item()
        })

    return batch_res

for idx, (_, row) in enumerate(tqdm(train_file.iterrows(), total=len(train_file))):
    try:
        if idx < prev_end:
            continue

        batch.append(row)

        if len(batch) == batch_size:
            results += process_batch(batch)
            batch = []

        if (idx+1) % save_interval == 0:

            pd.DataFrame(results).to_csv(autosave_path, mode='a', index=False, header=write_header)
            print(f"Autosaved after {idx + 1} rows")
            results = []  # Clear results to save memory
            write_header = False

    except Exception as e:
        print(f"Error processing row {row['Figure_path']}: {e}")
        continue

# Process any leftover rows
if batch:
    process_batch(batch)

# Final save
pd.DataFrame(results).to_csv('final_results.csv', index=False)
print("Finished full run. Saved final results.")
```
