```python
import torch
from transformers import Qwen2VLForConditionalGeneration, AutoProcessor

MODEL_PATH = local_dir + "/VQA_Disease_Diagnosis"  # or your preferred modality/task model

model = Qwen2VLForConditionalGeneration.from_pretrained(
    MODEL_PATH,
    torch_dtype=torch.bfloat16 if torch.cuda.is_available() else torch.float32,
    attn_implementation="eager",
    device_map="auto"
)

processor = AutoProcessor.from_pretrained(MODEL_PATH)
```

```python
from PIL import Image
from qwen_vl_utils import process_vision_info
import re

# Step 1: Get first sample


def generate_message(row):
    image_path = "/content/images/" + row["figure_path"]
    question = row["question"]
    choices = [row["Choice A"], row["Choice B"], row["Choice C"], row["Choice D"]]

    # Step 2: Format the multiple-choice question string
    formatted_question = (
        f"{question.strip()} "
        f"A) {choices[0]}, B) {choices[1]}, C) {choices[2]}, D) {choices[3]}. "
        "Please reason through the image and question carefully. "
        "Eliminate incorrect options one by one, and justify your reasoning with visual or clinical clues. "
        "Write your thought process in <think> </think> tags, and final answer (A, B, C, or D) in <answer> </answer>."
    )

    # Step 3: Load and resize the image
    image = Image.open(image_path).convert("RGB")
    image = image.resize((384, 384))
    # Wrap as expected by processor
    message = [{
        "role": "user",
        "content": [
            {"type": "image", "image": image},
            {"type": "text", "text": formatted_question}
        ]
    }]
    return message

def extract_reasoning_answer(text):
    #print(text)
    think_matches = [m.strip() for m in re.findall(r"<think>(.*?)</think>", text, re.DOTALL) if m.strip()]
    answer_matches = [m.strip() for m in re.findall(r"<answer>(.*?)</answer>", text, re.DOTALL) if m.strip()]
    #print(f"think: {think_match.group(1)}")
    #print(f"answer: {answer_match.group(1)}")

    reasoning = think_matches[-1] if think_matches else None
    answer = answer_matches[-1] if answer_matches else None

    return reasoning, answer
```

```python


import torch
from tqdm import tqdm

# ==== Configurable ====
batch_size = 16
save_count = 100
output_path = "partial_output.csv"
start_index = 11184
# ======================

all_outputs = []
skipped_rows = []

for i in tqdm(range(start_index, len(df), batch_size)):
    batch = df.iloc[i:i+batch_size]

    messages = []
    local_paths = []
    for _, row in batch.iterrows():
        try:
            message = generate_message(row)
            messages.append(message)
            local_paths.append(row["figure_path"])
        except Exception as e:
            print(f"Skipping row {row.name}: {e}")
            skipped_rows.append(row)
            continue

    if not messages:
        continue

    try:
        # Prepare batched inputs
        text = [processor.apply_chat_template(m, tokenize=False, add_generation_prompt=True) for m in messages]
        image_inputs, video_inputs = process_vision_info(messages)
        inputs = processor(text=text, images=image_inputs, videos=video_inputs, return_tensors="pt", padding=True).to("cuda")

        # Inference
        with torch.no_grad():
            generated_ids = model.generate(**inputs, use_cache=True, max_new_tokens=256)

        # Trim and decode
        generated_ids_trimmed = [
            out[len(inp):] for inp, out in zip(inputs.input_ids, generated_ids)
        ]
        batch_outputs = processor.batch_decode(
            generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=True
        )

        # Post-process and insert into DataFrame
        for j, (row_index, out_text) in enumerate(zip(batch.index, batch_outputs)):
            reasoning, answer = extract_reasoning_answer(out_text)
            df.at[row_index, "generated_reasoning"] = reasoning
            df.at[row_index, "generated_answer"] = answer
            df.at[row_index, "raw_output"] = out_text

    except Exception as e:
        print(f"Error in batch starting at index {i}: {e}")
        continue

    # Save periodically
    if (i + batch_size) % save_count < batch_size:
        df.iloc[:i+batch_size].to_csv(output_path, index=False)
        print(f"✅ Saved to {output_path} at row {i + batch_size}")

# Final save
df.to_csv(output_path, index=False)
print(f"✅ Final save completed to {output_path}")
```
